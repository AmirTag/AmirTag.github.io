```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f_SIS(x,t,alpha,beta):
    S, I = x
    return [-beta*I*S + alpha*I, beta*I*S - alpha*I]

I0 = 99
S0 = 1
N = I0 + S0
x0 = [S0, I0]

alpha = 0.5

contact_probability = 4/N
transmission_probability = 0.1
beta = contact_probability*transmission_probability

tf = 60
dt = 0.1
t = np.arange(0,tf,dt)

xt = odeint(f_SIS, x0,t,args=(alpha,beta))
```

```python
plt.figure(figsize=(12,8))

plt.suptitle(r"$r = %.2f$" %(N*beta-alpha), fontsize=24)
plt.fill_between(t,xt[:,1], xt[:,1] + xt[:,0],lw=2,label=r'$S(t)$')
plt.fill_between(t,np.zeros(len(t)),xt[:,1],lw=2,label=r'$I(t)$')
plt.xlabel("t (days)", fontsize=18)

plt.legend(fontsize=24)

plt.show()
```
